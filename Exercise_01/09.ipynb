```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme_03.png"

im = Image.open(my_path)
plt.imshow(im)
```

```python
# Grayscale convert
im_gray = im.convert("L")
plt.imshow(im_gray, cmap="gray")
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# Create a diagram
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title("Bieu do Histogram cua anh")
plt.xlabel("Gia tri Pixel")
plt.ylabel("So luong")
plt.show()
```

```python
# lam noi bat duong vien
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
