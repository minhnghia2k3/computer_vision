```python
import numpy as np
from PIL import Image

def average_images(image_list):
    first_img = Image.open(image_list[0])
    width, height = first_img.size
    total_array = np.array(first_img, 'f')
    count=1

    for image_path in image_list[1:]:
        try:
            im = Image.open(image_path)
            resized_im = im.resize((width, height))
            image_array = np.array(resized_im, 'f')
            total_array += image_array
            count=count+1
        except NameError:
            print(NameError)
            print("Skip ", image_path)

    # Tính trung bình
    average_array = total_array / count

    # Chuyển đổi kết quả trung bình thành hình ảnh
    average_image = Image.fromarray(average_array.astype('uint8'))

    return average_image

# Sử dụng ví dụ:
my_path_1 = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme.jpeg"
my_path_2 = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme_02.jpeg"
image_list = [my_path_1, my_path_2]
result_image = average_images(image_list)
display(result_image)


# Sử dụng ví dụ:
my_path_1 = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme_03.png"
my_path_2 = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme_04.png"
image_list = [my_path_1, my_path_2]
result_image = average_images(image_list)
display(result_image)
```
