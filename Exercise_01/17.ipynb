```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# Doc anh
img_path = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme.jpeg"
image = Image.open(img_path)

gray_image = image.convert("L")

image_array = np.array(gray_image)

# Tinh gradient theo huong X
gradient_x = ndimage.sobel(image_array, axis=0, mode='constant')

# Hien thi hinh anh
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2, 1, 2)
plt.imshow(gradient_x, cmap='gray')
plt.title("Anh gradient")

plt.show()
```

```python
# Doc anh
img_path = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme.jpeg"
image = Image.open(img_path)

gray_image = image.convert("L")

image_array = np.array(gray_image)

x_array = np.array([
    [0, 1, 1],
    [-1, 0, 1],
    [-1, 0, -1]
])

y_array = np.array([
    [-1, -1, -1],
    [0, 1, 1],
    [0, 1, 0]
])

# Tinh dao ham theo huong x
gradient_x = ndimage.convolve(image_array, x_array)

# Tinh dao ham theo huong y
gradient_y = ndimage.convolve(image_array, y_array)

# Tinh bien do theo gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Hien thi anh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Anh xam goc")

plt.subplot(2, 1, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title("Anh gradient")

plt.show()


```
