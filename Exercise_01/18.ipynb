```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image
```

```python
# Doc anh
image_path = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_stickers.jpeg"

image = Image.open(image_path)

# Chuyen anh sang mau xam
gray_image = image.convert("L")

# Chuyen anh sang array
image_array = np.array(gray_image)

# Chuyen sang anh nhi phan
binary_image = 1*(image_array<200)

# Label
labeled_image, num_features = ndimage.label(binary_image)

# Ve subplot
fig, axis = plt.subplots(3,1, figsize=(25, 25))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xam")

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhi phan")

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Da danh nhan ({num_features} doi tuong)")

plt.show();
```
