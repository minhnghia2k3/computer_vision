```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# Doc anh
img_path = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme_03.png"
image = Image.open(img_path)

# Anh xam'
gray_image = image.convert("L")

# Chuyen anh ve mang
image_a = np.array(gray_image)

# Làm mờ ảnh xám với bộ lọc Gaussian
blurred_image_a = ndimage.gaussian_filter(image_a, sigma=1, mode="constant")
blurred_image_b = ndimage.gaussian_filter(image_a, sigma=3, mode="constant")
blurred_image_c = ndimage.gaussian_filter(image_a, sigma=99, mode="constant")

# Hien thi anh
fig, axis = plt.subplots(2, 2, figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,2,2)
plt.imshow(blurred_image_a, cmap='gray')
plt.title('sigma=1')

plt.subplot(2,2,3)
plt.imshow(blurred_image_b, cmap='gray')
plt.title('sigma=3')

plt.subplot(2,2,4)
plt.imshow(blurred_image_c, cmap='gray')
plt.title('sigma=99')
```
