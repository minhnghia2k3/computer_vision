```python
my_path = "/Users/minhnghia2k3/Documents/Computer vision/Exercise_01/assets/cat_meme_02.jpeg"
im = Image.open(my_path)

# Convert image to array
im_array = np.array(im.convert("L"))

#
inverted_image = 255 - im_array
squared_image = im_array ** 2
clamped_image =np.clip(im_array, 100, 200)

# Show original image & transformed image
fig, axs = plt.subplots(2,2,figsize=(10,10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)

axs[0,0].imshow(im_array, cmap="gray")
axs[0,0].set_title("Original Image")

axs[0,1].imshow(inverted_image, cmap="gray")
axs[0, 1].set_title('Inverted Image')

axs[1,0].imshow(clamped_image, cmap="gray")
axs[1,0].set_title('Clamped Image')

axs[1,1].imshow(squared_image, cmap="gray")
axs[1,1].set_title('Squared Image')

plt.show()




```
